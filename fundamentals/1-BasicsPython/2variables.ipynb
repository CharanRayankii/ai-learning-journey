### Variables

A variable is like a container that stores data in memory. You can give it a name and use that name later in your program.

Example:
- `age = 25` stores the number 25
- `name = "Krish"` stores the text Krish

Variables make programs easier to read and work with.

```python
a=100
```

```python
## Declaring and Assigning Variables

age = 32
height = 6.1
name = "Krishna"
is_student = True

print("Age:", age)
print("Height:", height)
print("Name:", name)
print("Is Student:", is_student)
```

```python
## Naming Conventions

Variable names should be clear and easy to understand.

Rules:
- A name must start with a letter or `_`
- It can contain letters, numbers, and `_`
- Python is case-sensitive, so `name` and `Name` are different

Examples of valid names:

first_name = "Krish"
last_name = "Naik"
```

```python
# Invalid variable names
# 2age = 30          # starts with a number
# first-name = "Krish"  # contains a hyphen
# @name = "Krish"       # contains a special symbol
```

```python
## Case Sensitivity

Python treats uppercase and lowercase letters differently.

name = "Krish"
Name = "Naik"

print(name)
print(Name)
```

```python
## Understanding Variable Types

Python automatically detects the type of a variable based on the value assigned to it.

age = 25        # int
height = 6.1    # float
name = "KRish"  # str
is_student = True  # bool

print(type(name))
```

```python
## Type Checking and Conversion

height = 6.1
print(type(height))

height_str = str(height)
print(height_str)
print(type(height_str))
```

```python
age = 25
print(type(age))

# Convert integer to string
age_str = str(age)
print(age_str)
print(type(age_str))

# Convert string to integer
age_number = int("25")
print(age_number)
print(type(age_number))
```

```python
age = '25'
print(type(int(age)))
```

```python
## Dynamic Typing

Python allows a variable to change type during the program.

var = 10
print(var, type(var))

var = "Hello"
print(var, type(var))

var = 3.14
print(var, type(var))
```

```python
## Input from User

age = int(input("What is your age? "))
print("Your age is:", age)
print(type(age))
```

```python
### Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum_result = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
