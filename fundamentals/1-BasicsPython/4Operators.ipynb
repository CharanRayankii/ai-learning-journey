## Deep Dive into Operators

Operators are symbols that perform actions on values. In Python, they help you do math, compare values, and make decisions in your programs.

1. Introduction to Operators
2. Arithmetic Operators
   - Addition
   - Subtraction
   - Multiplication
   - Division
   - Floor Division
   - Modulus
   - Exponentiation
3. Comparison Operators
   - Equal to
   - Not equal to
   - Greater than
   - Less than
   - Greater than or equal to
   - Less than or equal to
4. Logical Operators
   - AND
   - OR
   - NOT
5. Practical Examples and Common Errors

```python
## Arithmetic Operations

# These operators are used for math calculations.
a = 10
b = 5

# Perform arithmetic operations
add_result = a + b
sub_result = a - b
mult_result = a * b
div_result = a / b
floor_div_result = a // b
modulus_result = a % b
exponent_result = a ** b

# Print the results
print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mult_result)
print("Division:", div_result)
print("Floor Division:", floor_div_result)
print("Modulus:", modulus_result)
print("Exponentiation:", exponent_result)
```

#### Comparison Operators

These operators compare two values and return either `True` or `False`.

```python
## Comparison Operators
## == Equal to

# Compare two values
a = 10
b = 10

print(a == b)
```

```python
# Compare two strings
str1 = "krishna"
str2 = "krishna"

print(str1 == str2)
```

```python
## Not Equal to !=

# Check if two values are different
str1 = "krishna"
str2 = "krishna"

print(str1 != str2)
```

```python
# Compare two strings with different cases
str3 = "krishna"
str4 = "krishna"

print(str3 != str4)
```

```python
# Greater than >

num1 = 45
num2 = 55

print(num1 > num2)
```

```python
## Less than <

# Check if one value is smaller than another
print(num1 < num2)
```

```python
# Greater than or equal to
number1 = 45
number2 = 45

print(number1 >= number2)
```

```python
# Less than or equal to
number1 = 44
number2 = 45

print(number1 <= number2)
```

#### Logical Operators

These operators help you combine conditions and make decisions in your code.

```python
## AND, OR, and NOT

# Logical operators combine conditions
x = True
y = True

result = x and y
print(result)
```

```python
# Another example using AND
x = False
y = True

result = x and y
print(result)
```

```python
## OR

# Example for OR logic
x = False
y = False

result = x or y
print(result)
```

```python
# NOT operator

x = False
print(not x)
```

```python
## Simple Calculator

# Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Performing arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2

# Displaying results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```
