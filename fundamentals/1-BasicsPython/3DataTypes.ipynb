## DataTypes

#### 1. Definition

- Data types tell Python what kind of value a variable stores.
- They help Python decide which operations are allowed on that value.
- They also help the program use memory more efficiently.

#### 2. Importance of Data Types in Programming

- Data types make programs easier to organize and understand.
- They help reduce errors by making sure values are used correctly.
- They allow Python to perform the right operations for each value.

```python
## Integer Example
age = 35
print(age)
print(type(age))
```

```python
## Floating Point Datatype
height = 5.11
print(height)
print(type(height))
```

```python
## String Datatype Example
name = "Krishna"
print(name)
print(type(name))
```

```python
## Boolean Datatype
is_true = True
print(is_true)
print(type(is_true))
```

```python
a = 10
b = 10

print(a == b)
print(type(a == b))
```

```python
## Common Error

# This will raise an error because Python cannot add a string and an integer directly
result = "Hello" + 5
```
