### Syntax and Semantics in Python

- Syntax means writing code in the correct format.
- Semantics means understanding what the code should do when it runs.

For example, `print("Hello")` has correct syntax, and when you run it, it displays the word Hello on the screen.

### Variables and Basic Rules

A variable stores data in Python. You can give it a name and assign a value to it. Python is case-sensitive, so `name` and `Name` are different variables.

```python
## Basic Syntax Rules in Python
## Python is case-sensitive

name = "Hare"
Name = "Krishna"

print(name)
print(Name)

# This shows that `name` and `Name` are two different variables.
```

### Indentation

Python uses indentation to show which statements belong to a block of code. This is very important in loops, conditions, and functions. If the indentation is wrong, Python will show an error.

```python
## Indentation Example
## Python uses spaces to group code inside an if statement

age = 32
if age > 30:
    print("Age is greater than 30")

print("This line is outside the if block")
```

### Comments

Comments are used to explain code. They help other people understand your program. In Python, single-line comments start with `#`, and multi-line comments are written inside triple quotes.

```python
## Line Continuation
## Use a backslash to continue a statement to the next line

total = 1 + 2 + 3 + \
    4 + 5 + 6

print(total)
```

```python
## Multiple Statements on One Line

Python allows more than one statement in one line, but it is better to keep code readable.

x = 5
y = 10
## Multiple Statements on One Line

Python allows more than one statement in one line, but it is better to keep code readable.

x = 5
y = 10
z = x + y
print(z)
print(z)
```

```python
## Understanding Semantics in Python

A variable stores data, and the value assigned to it gives it meaning.

age = 32       # age is an integer
name = "Krish" # name is a string

print(age)
print(name)
```

### Dynamically Typed

Python is a dynamically typed language. This means you do not need to declare the type of a variable before using it. Python automatically understands the type from the value you assign.

For example:
- `x = 10` → Python treats it as an integer
- `x = "Hello"` → Python treats it as a string

```python
## Type Inference Example

variable = 10
print(type(variable))

variable = "Krish"
print(type(variable))
```
